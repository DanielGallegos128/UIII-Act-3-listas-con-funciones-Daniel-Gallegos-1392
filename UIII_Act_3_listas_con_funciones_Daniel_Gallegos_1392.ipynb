```python
# Lista global para almacenar a los supervisores
supervisores = []

def agregarsupervisor(nombre):
    supervisores.append(nombre)
    print(f"\n✅ Supervisor '{nombre}' agregado correctamente.")

def leersupervisor(indice):
    try:
        if 0 <= indice < len(supervisores):
            print(f"\n🔍 Supervisor en índice {indice}: {supervisores[indice]}")
        else:
            print("\n❌ Error: Índice fuera de rango.")
    except ValueError:
        print("\n❌ Error: Por favor ingresa un número válido.")

def listarsupervisor():
    if not supervisores:
        print("\n📭 La lista está vacía.")
    else:
        print("\n--- 📋 Lista de Supervisores ---")
        for i, s in enumerate(supervisores):
            print(f"{i}. {s}")

def buscarsupervisor(nombre):
    if nombre in supervisores:
        pos = supervisores.index(nombre)
        print(f"\n📍 '{nombre}' encontrado en la posición {pos}.")
    else:
        print(f"\n❓ '{nombre}' no se encuentra en la lista.")

def modificarsupervisor(indice, nuevo_nombre):
    if 0 <= indice < len(supervisores):
        viejo_nombre = supervisores[indice]
        supervisores[indice] = nuevo_nombre
        print(f"\n🔄 Índice {indice} actualizado: '{viejo_nombre}' -> '{nuevo_nombre}'.")
    else:
        print("\n❌ Error: Índice inválido.")

def borrarsupervisor(nombre):
    if nombre in supervisores:
        supervisores.remove(nombre)
        print(f"\n🗑️ Supervisor '{nombre}' eliminado.")
    else:
        print(f"\n❌ Error: '{nombre}' no existe.")

# --- BLOQUE INTERACTIVO ---
def menu():
    while True:
        print("\n" + "="*30)
        print("   GESTIÓN DE SUPERVISORES")
        print("="*30)
        print("1. Agregar supervisor")
        print("2. Listar todos")
        print("3. Buscar por nombre")
        print("4. Modificar por índice")
        print("5. Borrar por nombre")
        print("6. Ver específico (índice)")
        print("7. Salir")

        opcion = input("\nSelecciona una opción (1-7): ")

        if opcion == "1":
            nom = input("Nombre del nuevo supervisor: ")
            agregarsupervisor(nom)

        elif opcion == "2":
            listarsupervisor()

        elif opcion == "3":
            nom = input("Nombre a buscar: ")
            buscarsupervisor(nom)

        elif opcion == "4":
            listarsupervisor() # Mostramos la lista para que vea los índices
            try:
                idx = int(input("Índice a modificar: "))
                nuevo = input("Nuevo nombre: ")
                modificarsupervisor(idx, nuevo)
            except ValueError:
                print("⚠️ Error: El índice debe ser un número.")

        elif opcion == "5":
            nom = input("Nombre a eliminar: ")
            borrarsupervisor(nom)

        elif opcion == "6":
            try:
                idx = int(input("Ingrese el índice: "))
                leersupervisor(idx)
            except ValueError:
                print("⚠️ Error: El índice debe ser un número.")

        elif opcion == "7":
            print("👋 Saliendo del sistema. ¡Hasta luego!")
            break

        else:
            print("🚫 Opción no válida, intenta de nuevo.")

# Ejecutar el menú
if __name__ == "__main__":
    menu()
```
